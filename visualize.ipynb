```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def plot(appResult, key, missingrate):
    records = []
    for model_idx, model in enumerate(key):
        for rate_idx, rate in enumerate(missingrate):
            dls, mae = appResult[model_idx][rate_idx]
            records.append({
                "Model": model,
                "Missing Rate": rate,
                "DLS": dls,
                "MAE": mae
            })

    df = pd.DataFrame(records)

    # Plotting DLS
    plt.figure(figsize=(10, 4))
    sns.lineplot(data=df, x="Missing Rate", y="DLS", hue="Model", marker="o")
    plt.title("Damerau-Levenshtein Similarity vs Missing Rate")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Plotting MAE
    plt.figure(figsize=(10, 4))
    sns.lineplot(data=df, x="Missing Rate", y="MAE", hue="Model", marker="o")
    plt.title("Remaining Runtime MAE vs Missing Rate")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
key = ['CDE', 'CRTP_LSTM', 'ED_LSTM', 'SuTraN']
missingrate = [0, 0.1, 0.3, 0.5, 0.7, 0.9]
addmissingrate = [0, 0.1, 0.3, 0.5, 0.7]
```
